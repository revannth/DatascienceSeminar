# Rename, Remove and Sorting in Pandas

```python
import pandas as pd
```

```python
marks=pd.read_csv('https://bit.ly/mtech-bangaloremarks')
```

```python
marks.head(5)
```

```python
cols = ['Rollno','State','Gender','Math','Read','Sc','Social %','Total']
marks.columns = cols
marks.head(5)
```

```python
marks.rename(columns={"Maths %":"Maths_%"})
```

```python
marks.columns = marks.columns.str.replace(' ','_' )
```

```python
marks.drop('Unnamed:_0',axis=1,inplace=True)
```

```python
marks.drop([0,1],axis=0)
```

```python
unsorted_data = pd.read_table('https://bit.ly/movieusers',sep='|',header=None,names=['Age','Gender','Occupation','Zipcode'])
```

```python
unsorted_data.head()
```

```python
unsorted_data.Age.sort_values()   #Sorting a Series
```

```python
unsorted_data.sort_values('Age')  #Sorting By a Series
```
