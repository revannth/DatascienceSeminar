# FILTERING, GROUPING AND STATISTICS

```python
import pandas as pd
```

```python
movies = pd.read_csv('https://bit.ly/mtech-imdbratings')
```

```python
movies.head()
```

```python
#Creating a list of booleans with the same exact length of a series
```

```python
booleans = []
for length in movies.duration:
    if length>200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
series_table = pd.Series(booleans)
```

```python
series_table.head()
```

```python
movies[series_table]
```

```python
movies[(movies.duration>=200)|(movies.genre=='Drama')]
```

```python
movies.groupby('genre').corr()  #When ever you want to analysis any pandas series by a category
```
