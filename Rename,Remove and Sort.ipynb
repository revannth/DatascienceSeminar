# Rename, Remove and Sorting in Pandas

```python
import pandas as pd
```

```python
marks=pd.read_csv('https://bit.ly/mtech-bangaloremarks')
```

```python
marks.head(5)
```

```python
cols = ['Rollno','State','Gender','Maths %','Reading %','Science %','Social %','Total']
marks.columns = marks.columns.str.replace(' ','_')
```

```python
marks.head(5)
```

```python
marks.rename(columns={"Maths %":"Maths_%"})
```

```python
marks.columns
```

```python
marks.drop([1],axis=0,inplace=True)
marks.head(5)
```

```python
marks.drop(['State','Gender','Maths_%'],axis=1)
marks.head(5)
```

```python
unsorted_data = pd.read_table('https://bit.ly/movieusers',sep='|',header=None,names=['Age','Gender','Occupation','Zipcode'])
```

```python
unsorted_data.head()
```

```python
unsorted_data.Gender.sort_values(ascending=False)   #Sorting a Series
```

```python
unsorted_data.sort_values(['Age','Gender'])  #Sorting By a Series
```
